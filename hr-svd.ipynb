```python
import numpy as np
import pandas as pd
```

```python
d_f = pd.read_csv('Book-01.csv')
```

```python
df = d_f[['CF_age band','Education Field','Marital Status','Job Role','CF_attrition count']]
```

```python
# import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import TruncatedSVD
```

```python
# preporcessing --> encoding 
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype.name == 'Category':
        df[col] = LabelEncoder().fit_transform(df[col])
```

```python
# standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
X = df[['CF_age band','Education Field','Marital Status','Job Role']]
y = df['CF_attrition count']
```

```python
# SVD represent documents and terms in vectors 
svd = TruncatedSVD(n_components=2, algorithm='randomized', n_iter=100, random_state=122)
X_svd = svd.fit_transform(X)

# plot the result
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y, s=50, cmap='viridis')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2-Component SVD')
plt.show()
```
