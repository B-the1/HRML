```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d_f = pd.read_csv('Book-01.csv')
```

```python
d_f.columns
```

```python
df = d_f[['Age','Job Level','Monthly Income', 'Performance Rating' ,'Total Working Years']]
```

```python
df.head()
```

```python
X = df.drop('Performance Rating',axis=1)
y=df['Performance Rating']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
mod = LinearRegression()
```

```python
mod.fit(X_train,y_train)
```

```python
print(mod.intercept_)
print(mod.coef_)
```

```python
y_prd = mod.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print('MSE ',mean_squared_error(y_test,y_prd))
print('R2s ',r2_score(y_test,y_prd))
print('RMSE ',np.sqrt(mean_squared_error(y_test,y_prd)))
print('MAE ',mean_absolute_error(y_test,y_prd))
```
