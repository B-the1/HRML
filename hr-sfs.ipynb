```python
import numpy as np
import pandas as pd
```

```python
d_f = pd.read_csv('Book-01.csv')
```

```python
df = d_f.select_dtypes(int)
```

```python
X  = df.drop(['Employee Number','CF_attrition rate','CF_current Employee','CF_attrition count'],axis=1)
y=df['CF_attrition count']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
sfs = SequentialFeatureSelector(model, n_features_to_select=11)
sfs = sfs.fit(X_train, y_train)
```

```python
selected_features = X.columns[sfs.get_support()]
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]
```

```python
selected_features
```

```python
model_base = LogisticRegression()
model_base.fit(X_train, y_train)
y_pred_base = model_base.predict(X_test)
accuracy_base = accuracy_score(y_test, y_pred_base)
print('\tAccuracy of the base model is', round(accuracy_base * 100, 2))
```

```python
model_sfs = LogisticRegression()
model_sfs.fit(X_train_selected, y_train)
y_pred_sfs = model_sfs.predict(X_test_selected)
accuracy_sfs = accuracy_score(y_test, y_pred_sfs)
print('\tAccuracy of the SFS-based model is', round(accuracy_sfs * 100, 2))
```
